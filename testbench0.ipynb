```python
import requests
from selenium import webdriver
import time
import validators

subject = 'math'

#base url
url = 'https://en.wikipedia.org/wiki/' + subject

print(url)

is_valid = False

validation = validators.url(url)

if validation:
  response = requests.get(url)
  print('URL: valid')
  if response.ok:
    print('Article: exists')
    is_valid = True
  else:
    print('Article: does not exist') 
else:
  print('URL: invalid')
```

```python
from bs4 import BeautifulSoup
import os

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
driver = webdriver.Chrome(options=chrome_options)

driver.get(url)

time.sleep(5)

html = driver.page_source

driver.get_screenshot_as_file("screenshot.png")
driver.quit()

soup = BeautifulSoup(html, 'html')

page_data = soup.prettify()

```

```python
with open('page_data.html', 'w') as f:
  f.write(page_data)
```

```python
anchors = []
links = []
topics = []
for a in soup.select('a[href^="/wiki/"]'):
    anchors.append(a)
    topics.append(str(a.string))
    links.append(a.get('href'))

for t in topics:
    print(t)

def remove_items(test_list, item):
    res = [i for i in test_list if i != item]
    return res

print(len(topics))
topics = remove_items(topics, 'None')
topics = remove_items(topics, 'Main page')
print(len(topics))
```

```python
# Prepare text
import nltk
nltk.download('punkt')
```

```python
from gensim.models import Word2Vec
from gensim.test.utils import common_texts

model = Word2Vec(sentences=common_texts, vector_size=100, window=5, min_count=10, workers=4)
model.save("word2vec.model")

model = Word2Vec.load("word2vec.model")
model.train(topics, total_examples=1, epochs=1)

```
